Q2. From the above results. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of 
Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" 
and “Product URL”. In case, if any of the details are missing for any of the product then 
replace it by “-“

```python
produrl =[]
produrl_ = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
```

```python
for i in produrl_:
    try: 
        produrl.append(i.get_attribute("href"))
    except NoSuchElementException:
        produrl.append('-')

   
```

```python
Brand = []
prod = []
rate = []
ratenum = []
price = []
ex = []
delv = []
avl = []
det = []
```

```python
for i in produrl:
    driver.get(i)
    time.sleep(5)
    try:
        brandtag = driver.find_element_by_xpath("//td[@class='a-size-base prodDetAttrValue']")
        Brand.append(brandtag.text)
    except:
        Brand.append('-')
    try:
        prodtag = driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
        prod.append(prodtag.text)
    except:
        prod.append('-')
    try:
        ratetag = driver.find_element_by_xpath("//div[@id='averageCustomerReviews']//span[@class='a-icon-alt']")
        rate.append(ratetag.get_attribute("innerHTML"))
    except NoSuchElementException:
        rate.append('-')
    try:
        ratenumtag = driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        ratenum.append(ratenumtag.get_attribute("innerHTML"))
    except:
        ratenum.append('-')
    try:
        pricetag = driver.find_element_by_xpath("//span[@id='priceblock_ourprice']")
        price.append(pricetag.text)
    except:
        price.append('-')
    
    try:
        extag = driver.find_element_by_xpath("//div[@class='a-accordion-row-a11y']//h5")
        ex.append(extag.text.replace('\n',' ').replace('   ',''))
    except:
        ex.append('-')
    
    try:
        delvtag = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b")
        delv.append(delvtag.text)
    except:
        delv.append('-')
    try:
        avltag = driver.find_element_by_xpath("//div[@id='availability']//span")
        avl.append(avltag.text)
    except:
        avl.append('-')
    try:
        dettag = driver.find_element_by_xpath("//div[@id='feature-bullets']")
        det.append(dettag.text.replace('\nShow More','.').replace('\n',''))
    except:
        det.append('-')
```

```python
Laptops = pd.DataFrame({})
Laptops['Brand Name'] = Brand[:10]
Laptops['Product Name']=prod[:10]
Laptops['Price']=price[:10]
Laptops['Rating']=rate[:10]
Laptops['Number of Ratings']=ratenum[:10] 
Laptops['Exchange']=ex[:10]
Laptops['Expected Delivery']=delv[:10]
Laptops['Availablitiy']=avl[:10]
Laptops['Details']=det[:10]


Laptops  
```

```python
Laptops.to_csv('Amazonscrape.csv')
```
